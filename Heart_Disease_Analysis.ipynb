```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\ms107\Downloads\Heart_new2 (1).csv")

df.columns = ['BMI','Smoking','AlcoholDrinking','Stroke','PhysicalHealth',
              'MentalHealth','DiffWalking','Sex','AgeCategory','Race',
              'Diabetic','PhysicalActivity','GenHealth','SleepTime',
              'Asthma','KidneyDisease','SkinCancer','HeartDisease']

print(df.head())
df.info()
print(df.isnull().sum())
```

```python
target = df.groupby('HeartDisease').size()
print(target)
```

```python
df['HeartDisease'].value_counts().plot(kind='bar', figsize=(8,6), color=["salmon","lightblue"])
plt.title("Heart Disease Distribution")
plt.show()
```

```python
hd = df.groupby('HeartDisease')['HeartDisease'].count()
plt.figure(figsize=(8,6))
plt.pie(hd, labels=['Absent','Present'], colors=['salmon','lightblue'], autopct='%0.0f%%')
plt.title('Heart Disease Population %')
plt.show()
```

```python
df.Sex.value_counts()
pd.crosstab(df.HeartDisease, df.Sex)

pd.crosstab(df.HeartDisease, df.Sex).plot(kind="bar",
                                          figsize=(10,6),
                                          color=["salmon","lightblue"])
plt.title("Heart Disease Frequency For Sex")
plt.xlabel("0 = Absent, 1 = Present")
plt.ylabel("Count")
plt.legend(["Female","Male"])
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='AgeCategory', hue='HeartDisease', data=df)
plt.title("Heart Disease by Age Category")
plt.xticks(rotation=45)
plt.show()
```
